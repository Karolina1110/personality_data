```python
import numpy as np
import pandas as pd
import utils
import seaborn as sn
```

# Load data

```python
data_big = pd.read_csv('data/big_results.csv', index_col=0)
data_big
```

# Drop null and reset index

```python
data_big=data_big.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
# Drop rows with NaN
data_big=data_big.reset_index(drop=True)
data_big
```

# Filter data by best classes

```python
best_class=data_big['TYPE'].value_counts()[:16].index.values.tolist()
best_class
```

```python
data_big_cleaned = data_big[data_big['TYPE'].isin(best_class)]
data_big_cleaned=data_big_cleaned.drop_duplicates()
data_big_cleaned=data_big_cleaned.reset_index(drop=True)
data_big_cleaned
```

```python
data_big_cleaned['TYPE'].value_counts()
```

# Add classes based on big 5 factors

extraversion - introversion E-I <br>
emotional stability - neurotic S-N <br>
agreeableness - antagonism A-T <br>
conscientiousness - carelessness C-R <br>
open - closed O-C 

```python
data_big_cleaned['B5_TYPE']=None
for index, row in data_big_cleaned.iterrows():
    t=''
    if row['EXT']>25:
        t+='E'
    else:
        t+='I'
    if row['EST']>25:
        t+='S'
    else:
        t+='N'
    if row['AGR']>25:
        t+='A'
    else:
        t+='T'
    if row['CSN']>25:
        t+='C'
    else:
        t+='R'
    if row['OPN']>25:
        t+='O'
    else:
        t+='C'
    data_big_cleaned.at[index, 'B5_TYPE']=t
    
data_big_cleaned
```

```python
print(len(data_big_cleaned['B5_TYPE'].unique()))
print(data_big_cleaned['B5_TYPE'].value_counts())
```

# Save new data

```python
data_big_cleaned.to_csv('data/big_cleaned_results.csv')  
```
