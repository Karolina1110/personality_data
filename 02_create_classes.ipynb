```python
import numpy as np
import pandas as pd
import utils
import seaborn as sn
import matplotlib.pyplot as plt
```

# Load clean data

```python
data = pd.read_csv('data/results_fix_notzero_final.csv', index_col=0)
data.head()
```

# Get sample of 500000 rows

```python
data_big = data.sample(n = 500000)
data_big
```

# Reset index

```python
data_big=data_big.reset_index(drop=True)
data_big
```

# Generate new classes

```python
def data_type_create(min_corr):
    
    # create new columns for personality type and correlation with other rows in this type
    data_big['TYPE']=None
    data_big['CORR']=None
    # create variable for type index
    i=0
    
    # choose columns for correlation calculation
    cols = list(range(50))
    cols.extend([100, 101, 102, 103, 104])
    
    # iterate over all rows
    for index1, row1 in data_big.iloc[:].iterrows():
        count=0
        # check if row has no type
        if data_big.iloc[index1]['TYPE']==None:
            print(i)
            # give row1 new type
            data_big.at[index1,'TYPE']=i
            
            # iterate over all rows
            for index2, row2 in data_big.iloc[:].iterrows():
                # calculate correlation between row1 and row2
                corr = np.corrcoef(row1.iloc[cols].tolist(), row2.iloc[cols].tolist())[0][1]
                
                # if correlation is bigger than given minimum and row2 has no type
                # or its previous correlation is smaller
                if (corr >= min_corr and data_big.iloc[index2]['CORR']==None) \
                or (corr >= min_corr and data_big.iloc[index2]['CORR'] < corr) :
                    # give row2 new type
                    data_big.at[index2,'TYPE']=i
                    # save correlation between rows
                    data_big.at[index2,'CORR']=corr
                    count+=1
            print('count:', count)
            # upgrade type index
            i=i+1
```

```python
data_type_create(0.95)
print(data['TYPE'].unique().size)
```
