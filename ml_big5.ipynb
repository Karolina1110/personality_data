```python
import numpy as np
import pandas as pd
import utils
import seaborn as sn
import matplotlib.pyplot as plt
from joblib import dump, load
```

```python
data = pd.read_csv('data/results.csv', index_col=0)
data.head()
```

```python
data=data.drop(['WINNER','CONFLICT'], axis=1)
data.head()
```

```python
data_big2 = data.copy()
data_big2=data_big2.dropna()
```

```python
data_big2['B5_TYPE']=None
for index, row in data_big2.iterrows():
    t=''
    if row['EXT']>25:
        t+='E'
    else:
        t+='I'
    if row['EST']>25:
        t+='S'
    else:
        t+='N'
    if row['AGR']>25:
        t+='A'
    else:
        t+='T'
    if row['CSN']>25:
        t+='C'
    else:
        t+='R'
    if row['OPN']>25:
        t+='O'
    else:
        t+='C'
    data_big2.at[index, 'B5_TYPE']=t
    
data_big2
```
