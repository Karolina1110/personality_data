```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn_som.som import SOM
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from joblib import dump, load
```

```python
data = pd.read_csv('data/results3.csv', index_col=0)
x = data.iloc[:, :50]
y = data.iloc[:, 106]
```

```python
le = LabelEncoder()
y_le = le.fit_transform(y)
le_dict = dict(zip(le.classes_,le.transform(le.classes_)))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_le, test_size=0.2, random_state=0)
```

```python
# som = load('data/som1.joblib')
som = SOM(m=5, n=1, dim=50, max_iter=10000)
som.fit(x_train.to_numpy())
predictions = som.predict(x_test.to_numpy())
```
