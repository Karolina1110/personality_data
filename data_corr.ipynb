```python
import numpy as np
import pandas as pd
import utils
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/results.csv', index_col=0)
data.head()
```

```python
data=data.drop(['WINNER','CONFLICT'], axis=1)
data.head()
```

```python
data['TYPE']=None
data['CORR']=None
data.head()
```

```python
data['TYPE']=None
data['CORR']=None
i=0
min_corr=0.70
for index1, row1 in data.iterrows():
    count=0
    if data.iloc[index1]['TYPE']==None:
#         print('index1:', i, index1)
        print(i)
        data.at[index1,'TYPE']=i
        for index2, row2 in data.iterrows():
#             print(np.corrcoef(row1.iloc[:50].tolist(), row2.iloc[:50].tolist())[0][1])
#             print('index2:', index2, np.corrcoef(data.iloc[index1,100:105].tolist(), data.iloc[index2,100:105].tolist())[0][1])
            corr = np.corrcoef(row1.iloc[100:105].tolist(), row2.iloc[100:105].tolist())[0][1]
            if (corr >= min_corr and data.iloc[index2]['CORR']==None) or (corr >= min_corr and data.iloc[index2]['CORR'] < corr) :
                data.at[index2,'TYPE']=i
                data.at[index2,'CORR']=corr
                count+=1
        i=i+1
        print('count:', count)
    else:
        pass
    
data.head()
```
