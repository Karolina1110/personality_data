```python
def data_type_create(min_corr):
    data_big['TYPE']=None
    data_big['CORR']=None
    i=0
    rows = list(range(50))
    rows.extend([100, 101, 102, 103, 104])
    for index1, row1 in data_big.iloc[:].iterrows():
        count=0
        if data_big.iloc[index1]['TYPE']==None:
    #         print('index1:', i, index1)
            print(i)
            data_big.at[index1,'TYPE']=i
            for index2, row2 in data_big.iloc[:].iterrows():
                corr = np.corrcoef(row1.iloc[rows].tolist(), row2.iloc[rows].tolist())[0][1]
                if (corr >= min_corr and data_big.iloc[index2]['CORR']==None) or (corr >= min_corr and data_big.iloc[index2]['CORR'] < corr) :
                    data_big.at[index2,'TYPE']=i
                    data_big.at[index2,'CORR']=corr
                    count+=1
            print('count:', count)
            i=i+1
        else:
            pass
#         if data['TYPE'].unique().size >= 22:
#             break

    data_big.head()
```

```python
data_type_create(0.95)
print(data['TYPE'].unique().size)
```
