```python
def data_type_create(min_corr):
    data['TYPE']=None
    data['CORR']=None
    i=1
    min_corr=0.9
    rows = list(range(50))
    rows.extend([100, 101, 102, 103, 104])
    for index1, row1 in data.iloc[:].iterrows():
        count=0
        if data.iloc[index1]['TYPE']==None:
    #         print('index1:', i, index1)
            print(i)
            data.at[index1,'TYPE']=i
            for index2, row2 in data.iloc[:].iterrows():
    #             print(np.corrcoef(row1.iloc[:50].tolist(), row2.iloc[:50].tolist())[0][1])
    #             print('index2:', index2, np.corrcoef(data.iloc[index1,100:105].tolist(), data.iloc[index2,100:105].tolist())[0][1])
                corr = np.corrcoef(row1.iloc[rows].tolist(), row2.iloc[rows].tolist())[0][1]
                if (corr >= min_corr and data.iloc[index2]['CORR']==None) or (corr >= min_corr and data.iloc[index2]['CORR'] < corr) :
                    data.at[index2,'TYPE']=i
                    data.at[index2,'CORR']=corr
                    count+=1
            print('count:', count)
            if count <= 5:
                for index3, row3 in data[data['TYPE']==i].iloc[:].iterrows():
                    data.at[index3,'TYPE']=0
            else:
                i=i+1
        else:
            pass
#         if data['TYPE'].unique().size >= 22:
#             break

    data.head()
```

```python
data_type_create(0.88)
corr=0.88
while data['TYPE'].unique().size >= 20:
    corr-=0.02
    data_type_create(corr)

print(data['TYPE'].unique().size)
print(corr)
```

```python
data_type_create(0.8)
print(data['TYPE'].unique().size)
```
